# **VISUALISATION**

```python
#Finding Correlation

corr = data.corr()
```

```python
# Visualising the correlation
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Finding lineplot for the correlation between energy for people as the loudness of the music increases
plt.figure(figsize=(12, 8))
sns.lineplot(corr, x = corr['loudness'], y = corr['energy']  , color = "c")
plt.xlabel("Loudness")
plt.ylabel("Energy")
plt.title("Correlation of Energy with Loudness ")
plt.show()
```

```python
# Creating a histogram for danceability
plt.figure(figsize=(10, 6))
sns.histplot(data['danceability'], bins=20, kde=True)
plt.xlabel('Danceability')
plt.ylabel('Frequency')
plt.title('Distribution of Danceability Scores')
plt.show()
```

```python
#Temporal Analysis of Trends
plt.figure(figsize=(12, 6))
sns.lineplot(x='weeks_on_chart', y='streams', data=data)
plt.xlabel('Weeks on Chart')
plt.ylabel('Streams')
plt.title('Popularity Trends Over Time')
plt.show()
```

```python
#Top Artists/Tracks by Streams
top_artists = data.groupby('artist_names')['streams'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_artists.index, y=top_artists.values, palette='inferno')
plt.xlabel('Artist')
plt.ylabel('Total Streams')
plt.title('Top 10 Artists by Total Streams')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#The most common keys and modes in the tracks
plt.figure(figsize=(10, 6))
sns.countplot(x='key', data=data, palette='viridis')
plt.xlabel('Key')
plt.ylabel('Count')
plt.title('Distribution of Keys')
plt.show()
```

```python
#Correlation between danceability and energy in the tracks
plt.figure(figsize=(10, 6))
sns.scatterplot(x='danceability', y='energy', data=data, )
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.title('Relationship Between Danceability and Energy')
plt.show()
```

```python
#Line Plot for Duration Trends Across Weeks on Chart
plt.figure(figsize=(12, 6))
sns.lineplot(x='weeks_on_chart', y='duration_ms', data=data, estimator=None, lw=1)
plt.xlabel('Weeks on Chart')
plt.ylabel('Duration (ms)')
plt.title('Duration Trends Across Weeks on Chart')
plt.show()
```

```python
#Swarm Plot for Energy Across Time Signatures
plt.figure(figsize=(10, 6))
sns.swarmplot(x='time_signature', y='energy', data=data)
plt.xlabel('Time Signature')
plt.ylabel('Energy')
plt.title('Energy Across Time Signatures')
plt.show()
```

```python
#Pair Plot for Multivariate Analysis
sns.pairplot(data[['danceability', 'energy', 'valence', 'loudness']])
plt.show()
```

```python
#Violin Plot for Danceability Distribution Across Keys
custom_palette = ['#FF5733', '#33FF7C', '#336BFF', '#FF33FF', '#FFDC33', '#A033FF', '#33FFDC', '#FF336B', '#33A6FF', '#6BFF33', '#FF33A6']
plt.figure(figsize=(10, 6))

# Annotate the violin plot with percentage values for each category
for i in range(len(data['key'].unique())):
    total_count = data['key'].value_counts().values[i]
    percentage = (total_count / len(data)) * 100
    plt.text(i, -0.1, f'{percentage:.1f}%', ha='center', va='center', color='black')

sns.violinplot(x='key', y='danceability', data=data, palette=custom_palette)
plt.xlabel('Key')
plt.ylabel('Danceability')
plt.title('Danceability Distribution Across Keys')
plt.show()
```
